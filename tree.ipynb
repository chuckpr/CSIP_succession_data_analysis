###Build and root a "phylum constrained" tree

```python
bt = parse_biom_table(open("data/otu_table_wtax.biom"))
phyla = {}
for oid, md in zip(bt.ObservationIds, bt.ObservationMetadata):
    try:
        p = md["taxonomy"][1]
    except IndexError:
        p = np.nan
    phyla[oid] = p
```

```python
phyla_s = pd.Series(phyla)
phyla_dummies = pd.get_dummies(phyla_s)
nrows, ncols = phyla_dummies.shape
def fast_tree_constraint(dummy_row):
    if dummy_row.sum(axis=1) == 1:
        return "".join([str(int(i)) for i in dummy_row.values])
    else:
        return "".join("-" for i in dummy_row.values)
constraint_s = phyla_dummies.apply(fast_tree_constraint, axis=1)
constraint_s.append(pd.Series(["".join("-" for i in range(ncols))], index=[sso_acc]))
constraint_s.to_csv("data/tmp/otusn_phylum_constraint_aln.txt",
                    sep = " ")
!echo $nrows $ncols | cat - data/tmp/otusn_phylum_constraint_aln.txt > data/tmp/otusn_phylum_constraint_aln.phy
!head data/tmp/otusn_phylum_constraint_aln.phy
```

```python
!FastTree -constraints data/tmp/otusn_phylum_constraint_aln.phy -gtr -nt data/aln_for_tree.fasta \
> data/otusn_phylum_constrained.tree 
```

```python
t_unroot = LoadTree("data/otusn_phylum_constrained.tree")
t_rooted = t_unroot.rootedWithTip(sso_acc)
t_rooted.writeToFile("data/otusn_phylum_constrained.tree")
```
