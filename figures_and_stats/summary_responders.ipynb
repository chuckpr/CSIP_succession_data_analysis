```python
%%R
library(dplyr)
library(tidyr)
library(ggplot2)
```

```python
%%R
l2fc.df = read.csv("../data/l2fc_table.csv")
copy.df = read.csv("../data/copynum.df")
shift.df = read.csv("../data/density_shifts.csv")
```

```python
%%R
FDR = 0.10

df.rspnse = l2fc.df %>%
    group_by(OTU, Treatment) %>%
    summarize(minP = min(padj)) %>%
    filter(minP <= FDR) %>%
    group_by() %>%
    spread(Treatment, minP)

rspnse = function(x) {
    ifelse(sum(x <= FDR, na.rm = TRUE) == 2, "both", ifelse(!is.na(x[1]), "Cellulose", "Xylose"))
}

df.rspnse$rspnse = apply(df.rspnse[,c("13CCPS", "13CXPS")], 1, rspnse)

df.rspnse = df.rspnse %>% select(OTU, rspnse)

df.rspnse
```

```python
%%R
FDR = 0.10
df.rspndr = l2fc.df  %>% filter(padj <= FDR)

df.rspndr = left_join(df.rspndr, copy.df)
df.rspndr = left_join(df.rspndr, df.rspnse)
df.rspndr = left_join(df.rspndr, shift.df)
```

```python
%%R
df.rspndr %>% select(Treatment, rspnse, Rank2, OTU, Day) %>% filter(Rank2 == "Proteobacteria")
```

```python
%%R -w 600 -h 600
d = df.rspndr %>% 
    select(OTU, Rank2, Treatment, log2FoldChange, shift, copynumber, Day, rspnse) %>%
    gather(key, value, -OTU, -Rank2, -Treatment, -Day, -rspnse)

p = ggplot(d, aes(x = Rank2, y = value))

p = p + facet_grid(key ~ Treatment, scales = "free", space = "free_x")

p = p + geom_violin()

p = p + geom_point(position = position_jitter(w = 0.25))

p = p + labs(x = "", y = "")

p = p + theme_bw()

p = p + theme(strip.background = element_blank(),
              strip.text = element_text(size = 14),
              axis.text.y = element_text(size = 14),
              axis.text.x = element_text(size = 14, face = "italic", hjust = 1, angle = 45))

p
```

```python
%%R -h 300 -w 600
d = df.rspndr %>% 
    select(OTU, Rank2, Treatment, log2FoldChange, shift, copynumber, Day, rspnse) %>%
    gather(key, value, -OTU, -Rank2, -Treatment, -Day, -rspnse) %>%
    filter(Rank2 == "Proteobacteria")

p = ggplot(d, aes(x = rspnse, y = value))

p = p + facet_wrap(~key, scales = "free_y")

p = p + geom_boxplot()

p = p + geom_point(na.rm = TRUE, position = position_jitter(w = 0.10)) 

p = p + theme_bw()

p = p + theme(strip.background = element_blank(),
              strip.text = element_text(size = 14),
              axis.text.y = element_text(size = 14),
              axis.text.x = element_text(size = 14, hjust = 1, angle = 45))

p
```
