```python
%%R
library(dplyr)
library(tidyr)
library(ggplot2)
```

```python
%%R
l2fc.df = read.csv("../data/l2fc_table.csv")
copy.df = read.csv("../data/copynum.df")
shift.df = read.csv("../data/density_shifts.csv")
```

```python
%%R
FDR = 0.10
df.rspndr = l2fc.df  %>% filter(padj <= FDR)

df.rspndr = left_join(df.rspndr, copy.df)
df.rspndr = left_join(df.rspndr, shift.df)
```

```python
%%R -w 600 -h 600
d = df.rspndr %>% 
    select(OTU, Rank2, Treatment, log2FoldChange, shift, copynumber, Day) %>%
    gather(key, value, -OTU, -Rank2, -Treatment, -Day)

p = ggplot(d, aes(x = Rank2, y = value))

p = p + facet_grid(key ~ Treatment, scales = "free", space = "free_x")

p = p + geom_violin()

p = p + geom_point(position = position_jitter(w = 0.25))

p = p + labs(x = "", y = "")

p = p + theme_bw()

p = p + theme(strip.background = element_blank(),
              strip.text = element_text(size = 14),
              axis.text.y = element_text(size = 14),
              axis.text.x = element_text(size = 14, face = "italic", hjust = 1, angle = 45))

p
```
