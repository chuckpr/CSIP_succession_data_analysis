```python
%%R
library(phyloseq)
library(ggplot2)
library(plyr); library(dplyr)
```

```python
%%R
physeq = import_biom("../data/otu_table_wtax.biom", "../data/otusn.tree")
sample.data = import_qiime_sample_data("../data/ABCD_combined_reads_mapper.tsv")
physeq = merge_phyloseq(physeq,sample.data)
physeq.md = sample_data(physeq)

physeq.PS.bulk = prune_samples((physeq.md$Treatment %in% c("12CCPS","13CCPS","13CXPS"))&
                               (row.names(physeq.md) %in% grep("bulk", row.names(physeq.md), value=TRUE)), physeq)   
physeq.PS.bulk = transform_sample_counts(physeq.PS.bulk, function(x) x/sum(x))
physeq.PS.bulk = filter_taxa(physeq.PS.bulk, function(x) sum(x > 0) > (0.25 * length(x)), TRUE)    
    
print(physeq.PS.bulk)
```

```python
%%R
mdf = psmelt(physeq.PS.bulk)
head(mdf)
```

```python
%%R
df.l2fc = read.csv("../data/l2fc_table.csv")
```

```python
%%R
mdf.byday = ddply(mdf, .(Day, OTU), summarize, mean_day_abundance = mean(Abundance))
mdf.byday = mdf.byday[order( mdf.byday$Day, -mdf.byday$mean_day_abundance),]
mdf.1 = subset(mdf.byday, Day == 1)
mdf.1$Rank = rank(-mdf.1$mean_day_abundance, ties.method = "random")
mdf.byday.ranked = join(mdf.byday, mdf.1[,c("OTU","Rank")], by = "OTU")
```

```python
%%R
FDR = 0.10

df.l2fc.rspndr = df.l2fc %>%
    group_by(OTU, Treatment) %>%
    summarize(minP = min(padj)) %>% 
    filter(minP <= FDR)
```

```python
%%R
FDR = 0.10

d.cell = left_join(df.l2fc.rspndr %>% filter(Treatment == "13CCPS"), mdf.byday.ranked)
d.xyl = left_join(df.l2fc.rspndr %>% filter(Treatment == "13CXPS"), mdf.byday.ranked)
```

```python
%%R -w 600 -h 350
p = ggplot(mdf.byday.ranked, aes(x = Rank, y = mean_day_abundance, color = factor(Day))) 

p = p + geom_vline(data = d.cell, aes(xintercept = Rank), color = "#377eb8", alpha = 0.25, size = 1)
p = p + geom_vline(data = d.xyl, aes(xintercept = Rank), color = "#4daf4a", alpha = 0.25, size = 1)

p = p + geom_line(alpha = 0.5) 

p = p + geom_rug(data = d.cell, color = "black", sides = "b", size = 1.25)
p = p + geom_rug(data = d.xyl, color = "black", sides = "t", size = 1.25)

p = p + geom_line(data = subset(mdf.byday.ranked, Day == 1), size = 2)

p = p + scale_y_log10()

p = p + scale_x_log10()

p = p + theme_bw()

p = p + labs(x = "Day 1 Rank", y = "Relative Abundance", color = "Day")

p
```

```python
%%R
ggsave("figs/RA_byday.pdf", p, h = 3.5, w = 6)
```

```python
%%R -w 350 -h 600

xyl.rspndrs = unique((df.l2fc.rspndr %>% filter(Treatment == "13CXPS"))$OTU)
d = mdf %>%
    filter(Rank2 %in% c("Actinobacteria", "Bacteroidetes", "Firmicutes", "Proteobacteria"), 
           OTU %in% xyl.rspndrs) %>%
    group_by(Day, OTU, Rank2) %>%
    summarize(Abundance = mean(Abundance))

p = ggplot(d, aes(x = factor(Day), y = Abundance))

p = p + facet_wrap(~Rank2, ncol = 1, scales = "free_y")

p = p + geom_bar(stat = "identity", fill = "black", color = "gray")

p = p + labs(x = "Day")

p = p + theme_bw()

p = p + theme(strip.background = element_blank(),
              strip.text = element_text(size = 14, face = "italic"),
              axis.text = element_text(size = 14),
              axis.title = element_text(size = 14))

p
```
