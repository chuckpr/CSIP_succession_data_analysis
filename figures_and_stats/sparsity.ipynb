```python
%%R
library(phyloseq)
library(dplyr)
library(ggplot2)
library(tidyr)
```

```python
%%R

asNumeric = function(x) { as.numeric(as.character(x)) }

physeq = import_biom("../data/otu_table_wtax.biom", "../data/otusn.tree")
sample.data = import_qiime_sample_data("../data/ABCD_combined_reads_mapper.tsv")
sample.data$Density = asNumeric(sample.data$Density)

physeq = merge_phyloseq(physeq,sample.data)

md = sample_data(physeq)
physeq = prune_samples(!(row.names(md) %in% grep("bulk", row.names(md), value=TRUE)), physeq)
physeq
```

```python
%%R
density.min = 1.7125
density.max = 1.755

md = sample_data(physeq)
physeq.heavy = prune_samples((md$Density >= density.min)&
                             (md$Density <= density.max), physeq)
```

```python
%%R
mdf.all = psmelt(physeq)
mdf.heavy = psmelt(physeq.heavy)
mdf.all$type = "all"
mdf.heavy$type = "heavy"
mdf.cmb = rbind(mdf.all, mdf.heavy)
```

```python
%%R
sparsity.df = mdf.cmb %>%
    group_by(OTU, Day, Treatment, type) %>%
    summarize(sparsity = sum(Abundance > 0) / length(Abundance))

sparsity.df
```

```python
%%R -w 550 -h 450
d = sparsity.df %>% group_by(add = FALSE) %>% spread(type, sparsity)

p = ggplot(d, aes(x = all, y = heavy))

p = p + geom_point(shape = 21, alpha = 0.15, position = position_jitter(h = 0.025, w = 0.025))

p = p + geom_hline(y = 0.60)

p = p + stat_smooth()

p = p + theme_bw()

p = p + theme(axis.text = element_text(size = 14),
              axis.title = element_text(size = 16))

p
```
