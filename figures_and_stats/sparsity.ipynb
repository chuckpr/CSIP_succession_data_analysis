```python
%%R
library(phyloseq)
library(plyr); library(dplyr)
library(ggplot2)
library(tidyr)
```

```python
%%R

asNumeric = function(x) { as.numeric(as.character(x)) }

physeq = import_biom("../data/otu_table_wtax.biom", "../data/otusn.tree")
sample.data = import_qiime_sample_data("../data/ABCD_combined_reads_mapper.tsv")
sample.data$Density = asNumeric(sample.data$Density)

physeq = merge_phyloseq(physeq,sample.data)

md = sample_data(physeq)
physeq = prune_samples(!(row.names(md) %in% grep("bulk", row.names(md), value=TRUE)), physeq)
physeq
```

```python
%%R
density.min = 1.7125
density.max = 1.755

md = sample_data(physeq)
physeq.heavy = prune_samples((md$Density >= density.min)&
                             (md$Density <= density.max), physeq)
```

```python
%%R
mdf.all = psmelt(physeq)
mdf.heavy = psmelt(physeq.heavy)
mdf.all$type = "all"
mdf.heavy$type = "heavy"
mdf.cmb = rbind(mdf.all, mdf.heavy)
```

```python
%%R
sparsity.df = mdf.cmb %>%
    filter(Treatment %in% c("13CCPS", "13CXPS", "12CCPS")) %>%
    group_by(OTU, Day, Treatment, type) %>%
    summarize(fraction_nonzeros = sum(Abundance > 0) / length(Abundance))

sparsity.df
```

```python
%%R -w 550 -h 450
d = sparsity.df %>% filter(fraction_nonzeros > 0) %>% group_by(add = FALSE) %>% spread(type, fraction_nonzeros)

p = ggplot(d, aes(x = all, y = heavy))

p = p + geom_point(size = 3, alpha = 0.15, position = position_jitter(h = 0.025, w = 0.025))

#p = p + geom_hline(y = 0.60, size = 1.5, color = "red")

p = p + theme_bw()

p = p + theme(axis.text = element_text(size = 14),
              axis.title = element_text(size = 16))

p
```

```python
%%R
l2fc.df = read.table("../data/l2fc_table.csv",
           sep = ",",
           header = TRUE)
colnames(l2fc.df)
```

```python
%%R

asNumeric = function(x) { as.numeric(as.character(x)) }

p = import_biom("../data/otu_table_wtax.biom", "../data/otusn.tree")
sample.data = import_qiime_sample_data("../data/ABCD_combined_reads_mapper.tsv")
sample.data$Density = asNumeric(sample.data$Density)

p = merge_phyloseq(p,sample.data)

md = sample_data(p)
physeq.bulk = prune_samples((row.names(md) %in% grep("bulk", row.names(md), value=TRUE)), p)
physeq.bulk
```

```python
%%R
mdf.bulk = psmelt(physeq.bulk)
```

```python
%%R
mdf.bulk = mdf.bulk %>%
    group_by(Sample) %>%
    mutate(A.rel = Abundance / sum(Abundance))
```

```python
%%R -w 1000
d = left_join(sparsity.df, mdf.bulk) %>% select(OTU, Day, Treatment, A.rel, type, fraction_nonzeros, Rank2)

p = ggplot(d, aes(y = A.rel, x = fraction_nonzeros))

p = p + geom_point(size = 4, alpha = 0.1, shape = 20, position = position_jitter(w = 0.02))

p = p + facet_wrap(~type)

p = p + scale_y_log10()

p = p + geom_vline(x = 0.6, size = 1.5, color = "red")

p = p + theme_bw()

p = p + theme(strip.text = element_text(size = 16),
              strip.background = element_blank(),
              axis.text = element_text(size = 16),
              axis.title = element_text(size = 16))

p
```

```python
%%R
d.bulk = mdf.bulk %>%
    group_by(Sample) %>%
    mutate(R = rank(desc(Abundance), ties.method = "random")) %>%
    select(Sample, R, OTU, Day, Treatment, Abundance) %>%
    group_by()

d.bulk
```

```python
%%R
sample.sparsity = mdf.cmb %>%
    group_by(OTU, Day, Treatment, type) %>%
    summarize(fraction_nonzeros = sum(Abundance > 0) / length(Abundance))

sample.sparsity
```

```python
%%R -w 1200 -h 900
d = left_join(d.bulk, sample.sparsity) %>% filter(type == "all")

p = ggplot(d, aes(x = R, y = Abundance, color = fraction_nonzeros))

p = p + facet_wrap(~Sample, scales = "free_y")

p = p + geom_line(size = 1, color = "gray")

p = p + geom_point(size = 2, shape = 21)

p = p + scale_y_log10()

p = p + scale_x_log10()

p = p + scale_color_gradientn(colours = rev(rainbow(3)))

p = p + theme_bw()

p = p + theme(strip.text = element_text(size = 16),
              strip.background = element_blank(),
              axis.text = element_text(size = 16),
              axis.title = element_text(size = 16),
              panel.border = element_rect(size = 2))

p
```

```python
%%R -w 1200 -h 900
d = left_join(d.bulk, sample.sparsity) %>% filter(type == "all", Abundance > 0)

p = ggplot(d, aes(x = R, y = fraction_nonzeros))

p = p + facet_wrap(~Sample, scales = "free_x")

p = p + geom_point(size = 2.5, alpha = 0.75)

p = p + scale_y_log10()

p = p + scale_x_log10()

p = p + geom_vline(x = 200, color = "red")

p = p + theme_bw()

p = p + theme(strip.text = element_text(size = 16),
              strip.background = element_blank(),
              axis.text = element_text(size = 16),
              axis.title = element_text(size = 16),
              panel.border = element_rect(size = 2))

p
```
