```python
%load_ext hierarchymagic
from hierarchymagic import run_dot
from IPython.display import Image, FileLink
```

```python
cmd = """
digraph G {

    graph [fontname="helvetica"];
    node [fontname="helvetica italic", fontsize=10];
    edge [penwidth=1];

    subgraph cluster_amendment {
    xylose [label=<<br></br><sup style="font-size:25%">13</sup>C-xylose<br></br><br></br>>, shape=rectangle,
            style="filled, rounded", fillcolor="#4daf4a", fontsize=12, color="white", fontname="helvetica"]
    cellulose [label=<<br></br><sup style="font-size:50%">13</sup>C-cellulose<br></br><br></br>>, shape=rectangle,
               style="filled, rounded", fillcolor="#377eb8", fontsize=12, color="white", fontname="helvitica"]
    label = <<br></br>Amendment>;
    style = "rounded, filled";
    }

    subgraph cluster_chloroflexi {
        Herpetosiphon[style=filled,fillcolor="#cbd5e8"];
        label = Chloroflexi;
        style = "rounded";
    }
    
    subgraph cluster_verruco {
        Spartobacteria[style=filled,fillcolor="#cbd5e8"];
        label = Verrucomicrobia;
        style = "rounded";
    }
    
    subgraph cluster_bacteroidetes {
        Sphingobacteria[style=filled,fillcolor="#cbd5e8"];
        Flavobacteria[style=filled,fillcolor="#cbd5e8"];
        Cytophaga[style=filled,fillcolor="#cbd5e8"];
        label = Bacteroidetes;
        style = "rounded";
    }
    
    subgraph cluster_actino {
        Micrococcales[style=filled,fillcolor="#cbd5e8"];
        Pseudonocardiales[style=filled,fillcolor="#cbd5e8"];
        label = Actinobacteria;
        style = "rounded";
    }
    
    subgraph cluster_firmicutes {
        Bacilli[style=filled,fillcolor="#cbd5e8"];
        label = Firmicutes;
        style = "rounded";
    }
    
    subgraph cluster_plancto {
        Plantomycetacia[style=filled,fillcolor="#cbd5e8"];
        label = Planctomycetes;
        style = "rounded";
    }
    
    subgraph cluster_proteo {
        Gammaproteobacteria[style=filled,fillcolor="#cbd5e8"];
        other[style=filled,fillcolor="#cbd5e8",label="Alpha|Beta|Delta|proteobacteria"];
        label = Proteobacteria;
        style = "rounded";
    }

    xylose -> Bacilli [color="#4daf4a"];
    xylose -> Gammaproteobacteria [color="#4daf4a"];
    xylose -> other [color="#4daf4a"];
    xylose -> Flavobacteria [color="#4daf4a"];
    xylose -> Sphingobacteria [color="#4daf4a"];
    xylose -> Micrococcales [color="#4daf4a"];

    cellulose -> Herpetosiphon [color="#377eb8"];
    cellulose -> Gammaproteobacteria [color="#377eb8"];
    cellulose -> other [color="#377eb8"];
    cellulose -> Spartobacteria [color="#377eb8"];
    cellulose -> Plantomycetacia [color="#377eb8"]
    cellulose -> Cytophaga [color="#377eb8"]
    cellulose -> Pseudonocardiales [color="#377eb8"]

    Bacilli -> Micrococcales;
    Flavobacteria -> Micrococcales;
    Flavobacteria -> other;
    other -> Flavobacteria;
    other -> Sphingobacteria;
    Gammaproteobacteria -> Flavobacteria;
    Gammaproteobacteria -> other;
    Sphingobacteria -> other;
    Gammaproteobacteria -> Sphingobacteria;
    Gammaproteobacteria -> Micrococcales;
    Sphingobacteria -> Micrococcales;
    other -> Micrococcales;
    Bacilli -> Flavobacteria;
    Bacilli -> Sphingobacteria;
    Bacilli -> other;

}
"""

f = "figs/foodweb.png"
with open(f, 'w') as f: f.write(run_dot(cmd))
Image("figs/foodweb.png")
```
