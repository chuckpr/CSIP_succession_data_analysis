```python
%%R
library(dplyr)
library(ggplot2)
library(RColorBrewer)
```

```python
%%R
df.l2fc = read.table("../data/l2fc_table.csv", 
                     sep = ",", 
                     header = TRUE)
```

```python
%%R
df.shifts = read.table("../data/density_shifts.csv", 
                       sep = ",", 
                       header = TRUE,
                       na.string = "")
colnames(df.shifts)
```

```python
%%R -h 350
FDR = 0.10

d  = left_join(df.l2fc, df.shifts)

p = ggplot(d, aes(x = log2FoldChange, y = shift))

p = p + geom_point(alpha = 0.20)

p = p + geom_point(data = d %>% filter(padj <= FDR), color = "red")

p = p + stat_smooth(method = "lm")

print(summary(lm(shift ~ log2FoldChange, data = d)))

p = p + theme_bw()

p = p + theme(axis.text = element_text(size = 14), 
              axis.title = element_text(size = 16))

p = p + labs(y = "Density shift, g/mL")

p
```

```python
%%R -h 600 -w 400
FDR = 0.10

d  = left_join(df.l2fc, df.shifts) %>% filter(padj <= FDR)

p = ggplot(d, aes(x = Rank2, y = shift))

p = p + facet_wrap(~Treatment, ncol = 1)

p = p + geom_boxplot(aes(group = Rank2))

p = p + geom_point(size = 3, alpha = 0.5, position = position_jitter(w = 0.10))

p = p + theme_bw()

p = p + theme(axis.text.y = element_text(size = 14), 
              axis.title = element_text(size = 16), 
              axis.text.x = element_text(size = 14, angle = 45, hjust = 1),
              strip.text = element_text(size = 16),
              strip.background = element_blank())

p = p + labs(y = "Density shift, g/mL")

p
```

```python
%%R -h 350 -w 400
FDR = 0.10

d  = left_join(df.l2fc, df.shifts) %>% filter(padj <= FDR)

p = ggplot(d, aes(x = Treatment, y = shift))

p = p + geom_boxplot()

p = p + geom_point(size = 3, shape = 21, alpha = 0.5, position = position_jitter(w = 0.10))

p = p + theme_bw()

p = p + theme(axis.text.y = element_text(size = 14), 
              axis.title = element_text(size = 16), 
              axis.text.x = element_text(size = 14))

p = p + labs(y = "Density shift, g/mL")

p
```
