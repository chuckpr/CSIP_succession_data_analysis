```python
mdf = com.load_data("mdf")
```

```python
relative_abundance = lambda x: x / x.sum()
mdf["RelativeAbundance"] = mdf.groupby("Sample")["Abundance"].transform(relative_abundance)
```

```python
def plot_horizon(mdf_otu, ax, A_max, nbands = 25, alpha = 0.05, color = sns.color_palette()[0]):
    A = mdf_otu["RelativeAbundance"]
    X = mdf_otu["Density"]
    
    band_width = A_max / nbands
    bands = np.arange(band_width, A_max + band_width, band_width)[::-1]
    
    for band in bands:
        Y = A.apply(lambda x: x - band if x - band > 0 else 0)
        A = A.apply(lambda x: band if x > band else x)
        ax.fill_between(X, Y, alpha = alpha, color = color)
    
    ax.set_ylim(bottom=0, top=band_width)
```

```python
f = plt.figure()
f.set_size_inches(16, 6)
gs = gridspec.GridSpec(3, 1)

treatments = ["13CCPS", "12CCPS"]
c = {"13CCPS" : "#377eb8",
     "12CCPS" : "#ff7f00"}

days = [3, 7, 14, 30]

for gsi, otu in enumerate(["OTU.4322", "OTU.6", "OTU.1"]):
    gs0 = gridspec.GridSpecFromSubplotSpec(2, 4, subplot_spec=gs[gsi, 0])
    
    for j, d in enumerate(days):
        mdf_otu = mdf[(mdf["OTU"]==otu)&(mdf["Day"]==d)&(mdf["Treatment"].isin(treatments))]
        A_max = mdf_otu["RelativeAbundance"].max()
        
        for i, t in enumerate(treatments):
            ax = plt.Subplot(f, gs0[i,j])
            ax.locator_params(axis="y", nbins=2)
            ax.set_xlim(1.66,1.76)
            f.add_subplot(ax)
            
            if i == 1 or j > 0:
                ax.set_yticklabels([])
            if gsi < 1 or i == 0:
                ax.set_xticklabels([])

            plot_horizon(mdf_otu[mdf["Treatment"]==t].sort("Density"), 
                         ax, 
                         A_max = A_max, 
                         nbands = 10, 
                         alpha = 0.20, 
                         color = c[t])
```
