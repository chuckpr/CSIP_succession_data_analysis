```python
from pandas.tools.plotting import radviz
import pandas as pd
import matplotlib.pyplot as plt
import seaborn; seaborn.set()
```

```python
%matplotlib inline
```

```python
df = pd.read_csv("../data/eco_metrics.csv")
df.columns
```

```python
fig = plt.figure(figsize = (10,10))
d = df
c = "Treatment"
kwds = {"s":75,
        "alpha":0.50}
ax = radviz(d[["log2FoldChangeWtime", "copynumber", "shift", "Rank", c]], c, **kwds)
```
