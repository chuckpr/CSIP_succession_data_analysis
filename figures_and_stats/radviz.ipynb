```python
from pandas.tools.plotting import radviz
import pandas as pd
import matplotlib.pyplot as plt
import seaborn; seaborn.set()
```

```python
import brewer2mpl
from brewer2mpl.qualitative import Set1
```

```python
%matplotlib inline
```

```python
df = pd.read_csv("../data/eco_metrics.csv")
df.columns
```

```python
def get_cmap(column, df, colors = Set1):
    cats = df[column].unique()
    try:
        return colors[len(cats)].mpl_colormap
    except KeyError:
        return colors[len(cats) + 1].mpl_colormap
```

```python
fig = plt.figure(figsize = (10,10))
d = df
c = "Treatment"

cmap = get_cmap(c, d)

kwds = {"s":75,
        "alpha":0.50}
ax = radviz(d[["log2FoldChangeWtime", "copynumber", "shift", "Rank", c]], c, colormap=cmap, **kwds)
```
