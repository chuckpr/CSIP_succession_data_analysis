```python
mdf = com.load_data("mdf")
df_l2fc = pd.read_csv("../data/l2fc_table.csv", sep=",")
```

```python
den_min = mdf.Density.min()
den_max = mdf.Density.max()
print den_min, den_max
```

```python
np.linspace(den_min, den_max, 5)
```

```python
treatment = "13CXPS"
days = [1,3,7,14,30]
FDR = 0.10
rspndrs = df_l2fc[(df_l2fc["padj"] <= FDR)&(df_l2fc["treatment"]==treatment)].OTU

matplotlib.rcParams["axes.facecolor"] = "white"
matplotlib.rcParams["axes.edgecolor"] = "#151515"
matplotlib.rcParams["grid.color"] = "#F5F5F5"
matplotlib.rcParams["grid.alpha"] = "0.75"
matplotlib.rcParams["ytick.labelsize"] = "x-small"
matplotlib.rcParams["xtick.labelsize"] = "x-small"
matplotlib.rcParams["axes.formatter.limits"] = (10,10)

N = len(rspndrs.unique())
#N = 2
fig, axarr = plt.subplots(ncols=len(days) + 1, nrows=N)

fig.set_size_inches(8, N * 1.25)

fig.subplots_adjust(wspace = 0.05, hspace = 0.45)

for j, otu in enumerate(rspndrs.unique()[:N]):
          
    for i, day in enumerate(days):
        
        d = mdf[(mdf["OTU"]==otu)&(mdf["Day"]==day)]
        no_bulk = d["Sample"].map(lambda x: not x.endswith("bulk"))
       
        d_treatment = d[(d["Treatment"]==treatment)&no_bulk]
        d_treatment["Density"] = d_treatment["Density"].astype(np.float)
        d_treatment.sort("Density", inplace=True)
        
        d_control = d[(d["Treatment"]=="12CCPS")&no_bulk]
        d_control["Density"] = d_control["Density"].astype(np.float)
        d_control.sort("Density", inplace=True)

        ax = axarr[j,i]
        ax.fill_between(d_treatment.Density.values, 
                        d_treatment.Abundance.values,
                        alpha = 0.5,
                        color = "#4daf4a")
        ax.fill_between(d_control.Density.values, 
                        d_control.Abundance.values,
                        color = "#ff7f00",
                        alpha = 0.5)
        ax.set_xticks(np.linspace(den_min, den_max, 5))
        ax.set_xticklabels([])
        ax.set_yticks([])

for j, otu in enumerate(rspndrs.unique()[:N]):

    d = mdf[mdf["OTU"]==otu]
    bulk = d["Sample"].map(lambda x: x.endswith("bulk"))
    d_bulk = d[bulk]
    
    i = len(days)
    ax = axarr[j,i]
    ax.set_xticks(days)
    ax.set_xticklabels([])
    ax.scatter(d_bulk[d_bulk.Treatment=="13CXPS"].Day.astype(int), 
               d_bulk[d_bulk.Treatment=="13CXPS"].Abundance.astype(float), 
               color = "#4daf4a",
               s = 10)
    ax.scatter(d_bulk[d_bulk.Treatment=="12CCPS"].Day.astype(int), 
               d_bulk[d_bulk.Treatment=="12CCPS"].Abundance.astype(float), 
               color = "#ff7f00",
               s = 10)
    ax.yaxis.tick_right()
    ax.autoscale(enable = True, axis = "y", tight = True)
 #   ax.autoscale(enable=True, tight=True)
    ax.set_ylim((0, ax.get_ylim()[1]))
    ax.set_yticks(np.linspace(ax.get_ylim()[0], ax.get_ylim()[1], 4))

    
    ymax = np.array([axarr[j,i].get_ylim()[1] for i in range(len(days))]).max()
    
    for i in range(len(days)):
        ax = axarr[j,i]
        ax.set_ylim((0, ymax))
    
    ax = axarr[j,1]
    ax.set_title(r"%s $%s$"%(otu, d.Rank2.iloc[0]), loc = 'left', fontsize = 8)

    ax = axarr[j,0]
    ax.set_yticks(np.linspace(ax.get_ylim()[0], 
                              ax.get_ylim()[1], 4))
    
for i in range(len(days)):
    ax = axarr[N-1, i]
    ax.set_xticklabels(ax.get_xticks(), rotation = 90, ha = "center")
    
ax = axarr[N-1, len(days)]
ax.set_xticklabels(days)
ax.set_xlabel("Day")
    
fig.text(0.005, 0.5, "Relative Abundance", rotation="vertical", va ="center", fontsize=16)
fig.text(0.35, 0.11 * N / len(rspndrs.unique()), r"Density, $g/mL$", fontsize=16)
```
