```python
%%R
library(ggplot2)
```

```python
%%R
df.l2fc = read.table("../data/l2fc.csv", sep=",", header = TRUE)
head(df.l2fc)
```

```python
%%R
unique(df.l2fc$Treatment)
```

```python
%%R -w 800 -h 600

FDR = 0.10 

colors = c("#377eb8","#4daf4a")

p = ggplot(df.l2fc, aes(x = log2FoldChange))

p = p + facet_wrap(~Rank2)

p = p + geom_density()

p = p + geom_rug()

p = p + geom_rug(data = subset(df.l2fc, (p.adj < FDR)), 
                 aes(color = Treatment), size = 1.5)

p = p + scale_color_manual(values = colors)

p = p + theme_bw()

#expression(paste("µg ", CO[2], " - C ", m^-2, " ", h^-1, sep="")))
p = p + labs(x = expression(paste(Log[2], " Fold Change")),
             y = "Kernel Density")

p
```

```python
%%R
ggsave("../figs/l2fc_KDE.pdf", p, h = 12, w = 16)
```
