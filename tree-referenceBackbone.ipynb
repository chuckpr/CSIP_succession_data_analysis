###Build and root the tree.

```python
tax_all = pd.read_csv("/var/seq_data/silva/Silva111_qiime_formatted/taxonomy/99_Silva_111_taxa_map_RDP_6_levels.txt",
            sep=";|\t",
            header=None,
            names=["acc","Kindom","Phylum","Class","Order","Family","Genus"])
tax_all.head()
```

```python
hits = !awk -F"\t" '{print $2}' data/tmp/silva111_blastout.m8
hits = pd.unique(hits)
tax_hits = tax_all[tax_all["acc"].isin(hits)]
dummies = pd.get_dummies(tax_hits["Phylum"]).apply(lambda x: "".join(str(int(i)) for i in x), axis=1)
tax_hits["tax_dummy"] = dummies.values
```

```python
nrows = tax_hits.shape[0]
ncols = len(tax_hits["tax_dummy"].values[0])
df = tax_hits[["acc","tax_dummy"]]
df.to_csv("data/tmp/constraint_aln.txt",
          sep=" ",
          index=False,
          header=False)
!head data/tmp/constraint_aln.txt
```

```python
!echo $nrows $ncols | cat - data/tmp/constraint_aln.txt > data/tmp/constraint_aln.phy
```

```python
!FastTree -constraints data/tmp/constraint_aln.phy -log data/tmp/silva111_hits.FT.log -gtr \
-nt data/tmp/silva111_hits.aln.fasta > \
data/tmp/silva111_hits.tree 
```

###Upload the tree to Itol to get a nice figure.

```python
itol_uploader = Itol.Itol()
url = itol_uploader.comm.upload_url
```

```python
bn = "data/tmp/silva111_hits.just_inserts"
r = requests.post(url, files={'treeFile': open(bn+".tree",'rb')})
tree_id = r.text.split()[1]
itol_exporter = ItolExport.ItolExport()
itol_exporter.add_export_param_dict({'tree':tree_id})
itol_exporter.add_export_param_dict({'format':'png'})
itol_exporter.export(bn+".png")
i = Image(filename=bn+".png", width=600)
display(i)
```
